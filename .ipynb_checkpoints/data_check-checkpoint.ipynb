```python
import pandas as pd
import duckdb
```

```python
file_path = '../fhvhv_tripdata_2026-01.parquet'
con = duckdb.connect()
```

```python
# --- Read with Pandas ---
df_pandas = pd.read_parquet(file_path)
# print("Shape:", df_pandas.shape)
df_pandas.head()
```

```python
df_duckdb = con.execute(f"SELECT count(1) FROM read_parquet('{file_path}') LIMIT 10").df()
df_duckdb
```

```python
df_duckdb = con.execute(f"SELECT * FROM read_parquet('{file_path}') LIMIT 10").df()
print("Shape (sample):", df_duckdb.shape)
df_duckdb
```

```python
df_duckdb.columns
```

```python
df_raw = con.execute(f"""
SELECT 
    hvfhs_license_num,
    dispatching_base_num,
    originating_base_num,
    request_datetime,
    pickup_datetime,
    dropoff_datetime,
    (pickup_datetime-request_datetime) as time_to_pickup,
    (dropoff_datetime - pickup_datetime) as time_to_arrival,
    trip_miles,
    trip_time,
    driver_pay
FROM read_parquet('{file_path}') 
LIMIT 10
""").df()
df_raw
```
